```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# graph sin and cos
from math import sin, cos
s = [sin(i * 0.1) for i in range(250)]
c = [cos(i * 0.1) for i in range(250)]
x = [i * 0.1 for i in range(250)]

fig, axs = plt.subplots(2)
axs[0].set_title("sin")
axs[0].plot(x, s)
axs[1].set_title("cos")
axs[1].plot(x, c);
```
